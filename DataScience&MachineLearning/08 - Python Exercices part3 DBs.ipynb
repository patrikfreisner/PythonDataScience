```python
# CENSUS_DATA
# CHICAGO_PUBLIC_SCHOOLS
# CHICAGO_CRIME_DATA

# Find the total number of crimes recorded in the CRIME table.
%sql select count(*) from CHICAGO_CRIME_DATA
```

```python
# List community areas with per capita income less than 11000.
%sql select community_area_name from census_data where per_capita_income < 11000
```

```python
# List all case numbers for crimes involving minors?
%sql select case_number from CHICAGO_CRIME_DATA where primary_type = 'OFFENSE INVOLVING CHILDREN'
```

```python
# List all kidnapping crimes involving a child?
%sql select case_number,primary_type from chicago_crime_data where primary_type = 'KIDNAPPING' and description like '%CHILD%'
```

```python
# What kind of crimes were recorded at schools?
%sql select s.name_of_school, c.primary_type, c.description from chicago_crime_data as c, chicago_public_schools as s where LOWER(c.location_description) like '%school%' and c.community_area_number = s.community_area_number
```

```python
# List the average safety score for all types of schools.
%sql select elementary_middle_or_high_school,AVG(safety_score) from chicago_public_schools group by elementary_middle_or_high_school
```

```python
# List 5 community areas with highest % of households below poverty line
%sql select community_area_name, percent_households_below_poverty from census_data order by percent_households_below_poverty desc LIMIT 5
```

```python
# Which community area(number) is most crime prone?
%sql select community_area_number, count(*) as Number_of_Crime from CHICAGO_CRIME_DATA where community_area_number is not NULL group by community_area_number order by number_of_crime desc limit 1
```

```python
# Use a sub-query to find the name of the community area with highest hardship index.
%sql select community_area_name from census_data where hardship_index in (select max(hardship_index) from census_data)
```

```python
# Use a sub-query to determine the Community Area Name with most number of crimes?
%sql SELECT COMMUNITY_AREA_NAME FROM CENSUS_DATA WHERE COMMUNITY_AREA_NUMBER IN (SELECT COMMUNITY_AREA_NUMBER from CHICAGO_CRIME_DATA where community_area_number is not NULL group by COMMUNITY_AREA_NUMBER order by COUNT(*) DESC LIMIT 1)
```
